```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt

def plot_simulation_data(target_dir="data"):
    if not os.path.exists(target_dir):
        print(f"Error: Directory '{target_dir}' not found.")
        print("Please run the simulation first or check the directory path.")
        return

    print(f"Scanning directory: {target_dir} ...\n")

    # 1. Spatial Spin Profile 그래프 그리기
    profile_files = glob.glob(os.path.join(target_dir, "spin_profile_*.txt"))
    if profile_files:
        plt.figure(figsize=(10, 6))
        for f in profile_files:
            filename = os.path.basename(f)
            label_name = filename.replace("spin_profile_", "").replace(".txt", "")
            data = np.loadtxt(f)
            # x_coord, spin_profile_mpi
            plt.plot(data[:, 0], data[:, 1], marker='o', markersize=3, linestyle='-', label=label_name)
        
        plt.title("Spatial Spin Profile $\langle s_i \rangle$")
        plt.xlabel("Position (x)")
        plt.ylabel("Magnetization $\langle s \rangle$")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.savefig(os.path.join(target_dir, "plot_spin_profile.png"), dpi=300)
        plt.close()
        print(" [✓] Saved: plot_spin_profile.png")

    # 2. Magnetization vs Time (MCS) 그래프 그리기
    mag_files = glob.glob(os.path.join(target_dir, "mag_time_*.txt"))
    if mag_files:
        plt.figure(figsize=(10, 6))
        for f in mag_files:
            filename = os.path.basename(f)
            label_name = filename.replace("mag_time_", "").replace(".txt", "")
            data = np.loadtxt(f)
            # j (MCS), mag_time_mpi
            plt.plot(data[:, 0], data[:, 1], linestyle='-', linewidth=1.5, label=label_name)
        
        plt.title("Total Magnetization vs Monte Carlo Steps")
        plt.xlabel("Monte Carlo Steps (MCS)")
        plt.ylabel("Total Magnetization (m)")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.savefig(os.path.join(target_dir, "plot_mag_time.png"), dpi=300)
        plt.close()
        print(" [✓] Saved: plot_mag_time.png")

    # 3. Spin Correlation Function 시간별 스냅샷 그래프 그리기
    corr_files = glob.glob(os.path.join(target_dir, "spin_correlation_*.txt"))
    for f in corr_files:
        filename = os.path.basename(f)
        
        with open(f, 'r') as file:
            header = file.readline().strip().split()
        
        data = np.loadtxt(f, skiprows=1)
        r = data[:, 0]
        
        plt.figure(figsize=(10, 6))
        # header[0]은 'r', header[1:]은 'MCS_10', 'MCS_20' 등
        for i, label in enumerate(header[1:]):
            plt.plot(r, data[:, i+1], linestyle='-', marker='', label=label)
            
        plt.title(f"Spin Correlation Function\n({filename.replace('.txt', '')})")
        plt.xlabel("Distance (r)")
        plt.ylabel("Correlation $\langle s_i s_{i+r} \rangle$")
        # x축을 로그 스케일로 보고 싶다면 아래 주석을 해제하세요.
        # plt.xscale('log') 
        plt.grid(True, linestyle='--', alpha=0.7)
        
        # 범례가 너무 많으면 그래프를 가리므로 바깥으로 빼줍니다.
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2, fontsize='small')
        plt.tight_layout()
        
        out_name = filename.replace("spin_correlation_", "plot_corr_").replace(".txt", ".png")
        plt.savefig(os.path.join(target_dir, out_name), dpi=300)
        plt.close()
        print(f" [✓] Saved: {out_name}")

if __name__ == "__main__":
    # 기본적으로 현재 디렉토리의 'data' 폴더를 탐색합니다.
    # 만약 multi_qsub.sh로 생성한 'results/T_...' 폴더를 탐색하고 싶다면 경로를 변경하세요.
    target_directory = "data" 
    plot_simulation_data(target_directory)
```
